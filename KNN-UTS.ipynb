```python
from sklearn.preprocessing import LabelEncoder                                              #Muamar Dwihanggoro 202131036
```

```python
#mengubah values kolom yang bertipe data string/object menjadi numerik
for col in ['job','marital','education','default','housing','loan','contact','month','poutcome','y']:
    label_encoder = LabelEncoder()
    dataset[col] = label_encoder.fit_transform(dataset[col])
```

```python
dataset
```

```python
x = dataset.drop(["job","marital","education","default","housing","loan","contact","month","poutcome","y"], axis=1)
y = dataset["y"]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train, y_train)
```

```python
print("Akurasi Model Data : {}".format(knn.score(x_test,y_test)))
```

```python
import numpy as np
x_new = np.array ([[43,668,5,508,5,-1,1]])
prediksi = knn.predict(x_new)
print(prediksi)
```
